b) Remove the units from some of the attributes and only keep the numerical values (for
example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from
“New_price”). (4 points)

```python
data['Mileage'] = data['Mileage'].str.extract(r'(\d+\.?\d*)')[0].astype(float)
data['Engine'] = data['Engine'].str.replace(' CC', '', regex=False).astype(float)
data['Power'] = data['Power'].str.replace(' bhp', '', regex=False).astype(float)
```

```python
data['New_Price'] = pd.to_numeric(data['New_Price'].str.replace(' lakh', '', regex=False), errors='coerce')
```

```python
data[['Mileage', 'Engine', 'Power', 'New_Price']].head(15)
```

```python
# Save the updated dataset to a CSV file
data.to_csv('afterTask2.csv', index=False)
```

```python
# Define mileage range
ranges = [0, 15, 20, 25, float('inf')]
labels = ['Low', 'Medium', 'High', 'Very High']

data['Mileage_Category'] = pd.cut(data['Mileage'], bins=ranges, labels=labels)
data[['Mileage','Mileage_Category']].head(10)
```

e) Perform select, filter, rename, mutate, arrange and summarize with group by operations (or
their equivalent operations in python) on this dataset. (4 points)

```python
cols = data[['Name', 'Year', 'Mileage', 'Price','Owner_Type','Price','Seats','Power']] #select operation
cols
```

```python
rows_filetered = data[(data['Owner_Type'] == 'First') & (data['Price'] > 25)]#Filtering cars that are 'First' owner type and have a price greater than 25
rows_filetered
```

```python
cols = data.rename(columns=lambda x: {'Power': 'Horsepower', 'Seats': 'Num_Seats'}.get(x, x))
cols
```

```python
cols['Horsepower'] = cols['Horsepower'] * 1.34102  # 1 kW = 1.34102 BHP
cols['Horsepower']
```

```python
cols = cols.sort_values(by=['Mileage', 'Year'], ascending=[False, True]) #Arrange by Mileage (descending) and Year (ascending)
cols[['Mileage','Year']]
```

```python
# Group by 'Owner_Type' and calculating  mean price and mileage
summarized_data = cols.groupby('Owner_Type').agg(
    Average_Price=('Price', 'mean'),
    Average_Mileage=('Mileage', 'mean')
).reset_index()
summarized_data
```

```python
# Save the updated dataset to a CSV file
cols.to_csv('Operations.csv', index=False)
```
